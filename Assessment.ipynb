```python
plt.figure(figsize=(14,10))

sns.set(style='darkgrid')
```

```python
sns.barplot(x=datak.fuel_type_code_pudl,y=datak.fuel_qty_burned,color='#69b3a2')
plt.xticks(rotation=90)
plt.xlabel('quantity of fuel type burned')
```

```python
sns.boxplot(x=datak.fuel_type_code_pudl,y=datak.fuel_qty_burned,color='#69b3a2')
plt.xticks(rotation=90)
plt.xlabel('fuel quantity burned per year')
```

```python
sns.violinplot(x=datak.report_year, y=datak.fuel_qty_burned)
plt.xticks(rotation=90)
plt.xlabel('fuel quantity burned per year')
```

```python
sns.barplot(x=datak.report_year, y=datak.fuel_qty_burned)
plt.xticks(rotation=90)
plt.xlabel('fuel quantity burned per year')
```
